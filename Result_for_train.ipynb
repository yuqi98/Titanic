```python
import csv
with open('train.csv', 'r') as f:
    reader = csv.reader(f)
    train_list = list(reader)
    
```

```python
length = len(train_list)
print(length)
```

```python
import random
result_list = [["PassengerId","Survived"]]
for i in range(1,length):
    if train_list[i][2] == "1" and train_list[i][4] == "female" :
        result_list = result_list + [[i,'1']]
    
    if train_list[i][2] == "1" and train_list[i][4] == "male" :
        a = random.randint(0,1)
        if a == 1:
            result_list = result_list + [[i,'1']]
        else:
            result_list = result_list + [[i,'0']]
    if train_list[i][2] != "1":
        result_list = result_list + [[i,'0']]
            
print(result_list)
    
```

```python
import pandas as pd 
my_df = pd.DataFrame(result_list)
my_df.to_csv('train_out.csv', index=False, header=False)
print(my_df)
```
