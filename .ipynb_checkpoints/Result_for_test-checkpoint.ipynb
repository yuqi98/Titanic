```python
import csv
with open('test.csv', 'r') as f:
    reader = csv.reader(f)
    test_list = list(reader)
```

```python
length = len(test_list)
print(length)
```

```python
import random
result_list = [["PassengerId","Survived"]]
for i in range(1,length):
    if test_list[i][1] == "1" and test_list[i][3] == "female" :
        result_list = result_list + [[i,'1']]
    
    if test_list[i][1] == "1" and test_list[i][3] == "male" :
        a = random.randint(0,1)
        if a == 1:
            result_list = result_list + [[i,'1']]
        else:
            result_list = result_list + [[i,'0']]
    if test_list[i][1] != "1":
        result_list = result_list + [[i,'0']]
            
print(result_list)
    
```

```python
import pandas as pd 
my_df = pd.DataFrame(result_list)
my_df.to_csv('test_out.csv', index=False, header=False)
print(my_df)
```
